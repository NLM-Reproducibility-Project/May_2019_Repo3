First we will create a data directory if it doesn't already exist and enter it

```bash
cd ..
mkdir -p data
cd data
```

1. Next we will download the data from the Schmid et al., 2018 using SRAtools fastq-dump
    Accessions are saved in data/30157438_accessionlist.txt

```bash
while IFS='' read -r line || [[ -n "$line" ]]; 
    do
        if [ ! -f "$line.fastq.gz" ]; then
        fastq-dump --gzip $line;
        fi
    done < 30157438_accessionlist.txt
```
